```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('flight_delays.csv', encoding='latin1')
```

```python
df = pd.read_csv('flight_delays.csv')
plt.figure(figsize=(10, 6))
plt.bar(df['Month'], df['NK'])
plt.title("Average Arrival Delay for Spirit Airlines Flights, by Months")
plt.xlabel("Month")
plt.ylabel("Arrival delay (in minutes)")
plt.bar(df['Month'], df['NK'], color='blue', width=0.5)
plt.show()
```

```python
df = pd.read_csv('taxi_trip.csv')
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['trip_duration'], df['tip'], color='blue', linestyle='-', linewidth=2, marker='o')
plt.title("Trip Duration vs Tip Received")
plt.xlabel("Trip Duration")
plt.ylabel("Tip Received (in USD)")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['trip_duration'], df['total_fare'], color='red', linestyle='--', linewidth=2, marker='s', label='Total Fare')
plt.plot(df['trip_duration'], df['tip'], color='blue', linestyle='-', linewidth=2, marker='o', label='Tip Received')
plt.title("Total Fare and Tips Received by Trip Duration")
plt.xlabel("Trip Duration")
plt.ylabel("Amount (in USD)")
plt.legend()
plt.show()
```
